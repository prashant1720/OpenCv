# 1.Gaussian Blur

```python
img_g = cv2.GaussianBlur(img,(15,15),5) #use 15*15 Gaussian 
plt.imshow(img_g[...,::-1])
plt.show()
```

# conversion to grayscale

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_gray[...,:],cmap='gray')
plt.show()
```

# image Blending

```python
img_rotate = img[::-1,...]
img_blend= img_rotate/2+img/2
plt.imshow(img_blend[...,::-1])
plt.show()
```

# EdgeDetection

```python
img_Canny = cv2.Canny(img,180,180)
plt.imshow(img_Canny[...,::-1],cmap='gray')
```

# opencv for camera-video

```python
cap = cv2.VideoCapture(0)# open camera with device 0 which is webcam
r,img= cap.read()
while r:
    cv2.imshow('Frame',img)
    a=cv2.waitKey(40)
    if a!=-1:  # if any key is pressed
        r=False # set r to False stop reading Frame
    else:
        r,img=cap.read() # else read new frame
cv2.destroyAllWindows()
cap.release()
```
